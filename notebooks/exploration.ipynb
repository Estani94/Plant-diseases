```python
import os
from pathlib import Path
from collections import Counter
from PIL import Image

base_path = Path("../data/raw")
classes = [folder.name for folder in base_path.iterdir() if folder.is_dir()]
class_counts = {cls: len(list((base_path / cls).glob("*"))) for cls in classes}

print("Clases:", classes)
print("Cantidad por clase:", class_counts)

sample_image_path = next((base_path / classes[0]).rglob("*.*"))
img = Image.open(sample_image_path)
print("Tamaño de muestra:", img.size)
```

```python
subfolder_path = base_path / "PlantVillage"
subclasses = [folder.name for folder in subfolder_path.iterdir() if folder.is_dir()]
subclass_counts = {cls: len(list((subfolder_path / cls).glob("*"))) for cls in subclasses}

print("Subclases:", subclasses)
print("Cantidad por subclase:", subclass_counts)
```

```python
import matplotlib.pyplot as plt
import random
from PIL import Image

fig, axes = plt.subplots(3, 5, figsize=(15, 9))
samples = random.sample(subclasses, 15)

for ax, subclass in zip(axes.flatten(), samples):
    img_path = random.choice(list((subfolder_path / subclass).rglob("*.*")))
    img = Image.open(img_path)
    ax.imshow(img)
    ax.set_title(subclass)
    ax.axis('off')

plt.tight_layout()
plt.show()
```
